# Gr√°ficas en matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Graficas en dos dimensiones

```python
x=np.array([1, 2, 3, 4, 5])
y=np.array([10, 5, 8, 15, 20])
```

```python
plt.plot(x,y,'*-r')
plt.xlabel('x')
plt.ylabel('y')
plt.grid('on')
plt.title('Grafica 1')
```

## Varias graficas en una figura

```python
x=np.linspace(1,5,10)
plt.subplot(1,2,1)
y1=x*(np.sin(x)+1)
plt.plot(x,y1,'r')
plt.title('y=x(sen(x)+1)')
plt.subplot(1,2,2)
y2=np.sin(x/2)/x
plt.plot(x,y2,'g')
plt.title('y=sen(x/2) / x')
plt.suptitle('Grafica 2')
```

```python
fig,ax = plt.subplots(2,2,figsize=(6,5))
ax[0,0].plot(x,y1,'r')
ax[0,0].set_title('y=x(sen(x)+1)')
ax[0,1].plot(x,y2,'g')
ax[0,1].set_title('y=sen(x/2) / x')
fig.suptitle('Grafica 2')
```

## Grafica de contornos

```python
dx, dy = 0.05, 0.05
y, x = np.mgrid[slice(1, 5 + dy, dy),
                slice(1, 5 + dx, dx)]

z = np.sin(x)**10 + np.cos(10 + y*x) * np.cos(x)
cmap = plt.get_cmap('PiYG')

fig, (ax0, ax1) = plt.subplots(nrows=2)

im = ax0.pcolormesh(x, y, z, cmap=cmap)
fig.colorbar(im, ax=ax0)
ax0.set_title('pcolormesh')

cf = ax1.contourf(x,y,z,cmap=cmap)
fig.colorbar(cf, ax=ax1)
ax1.set_title('contourf')

# ajusta el espacio entre graficas
fig.tight_layout()
```

## Gr√°ficas en tres dimensiones

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
